```python
import pandas as pd
import glob
import os

BASE_INPUT_DATA_PATH='../data/raw/'
BASE_OUTPUT_DATA_PATH='../data/combined/'
```

```python
csv_files=glob.glob(os.path.join(BASE_INPUT_DATA_PATH,"*.csv"))
dfs=[]
csv_files
```

```python
for f in csv_files:
    bank_name=os.path.basename(f).split(".")[0]
    df=pd.read_csv(f)
    df.columns=df.columns.str.strip()

    df['bank']=bank_name  # Add New Column
    df=df.rename(columns={'published_date':'date','traded_quantity':'volume','traded_amount':'amount'})
    
    df['date']=pd.to_datetime(df['date'])

    dfs.append(df)

combined_df=pd.concat(dfs,ignore_index=True)
combined_df=combined_df.loc[:,~combined_df.columns.str.contains('^Unnamed')]
combined_df.drop_duplicates(inplace=True)
combined_df=combined_df.sort_values(['bank','date'],ascending=[True,True])
combined_df=combined_df.drop('status',axis=1)
combined_df.to_csv(BASE_OUTPUT_DATA_PATH+'combinedCommercialBankData.csv',index=False)

```
